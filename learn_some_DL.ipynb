```python
import torch
from torch import nn
from torch.nn import functional as F
## 通过nn.Moule可以个性化定义，相比于
class MyModel(nn.Module):
    def __init__(self):
        self.first = nn.Linear(18,20)
        self.second = nn.Linear(20,18)
    
    def forward(self,X):
        return  self.second(F.relu(self.first(X)))
## MyModel可以自己定义，相比nn.Sequential()更多个性化。合起来用也很好
X = torch.rand(2,18)
# MyModel(X) == 
# net = nn.Sequential(nn.Linear(18,20),nn.ReLU,nn.Linear(20,18)) net(X)

## 同样可以用module，class定义每层
```
